# Null Hypothesis & Alternate Hypothesis

    1. Null Hypothesis (H0)
    
    the belief we consider, we can say thumb rule for the reference. base universal truth case
    
    e.g. No one will charge for crime if s/he's innocent(universal rule)
    
    ***so, we here in null hypothesis we will consider if one is not guilty(innocent) one will not be punished.
    
    criminal 
    H0: He is Innocent.
    h1: He is guilty.
    
      
    
    
    
    
    
    
    2. Alternate Hypothesis  H1 (opp. of the Null Hypothesis)
    
    e.g. Null hypothesis(H0)=> is fruit is mango
         Alternate Hypothesis (h1) => fruit is NOT EQUAL to mango
    
    *** we always say we failed to reject the null hypothesis,instead of accept the null hypothesis

# Rules

    H0: only = or ≥ or ≤ 
    h1: only ≠ or > or <
