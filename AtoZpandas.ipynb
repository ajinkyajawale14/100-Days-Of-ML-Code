# Merging & Concate Dataset

```python
d1=pd.DataFrame([['a',1],['b',2]],columns=['coll','number'])  
```

```python
d2=pd.DataFrame([['c',3,'lion'],['d',4,'tiger']],columns=['letter','number','animal'])
```

```python
d1
```

```python
d2
```

```python
#concate row wise
pd.concat([d1,d2],axis=0)
```

```python
#to remove index
pd.concat([d1,d2],axis=0,ignore_index=True,sort=True)
```

```python
pd.concat([d1,d2],axis=1,sort=True)
```

# Merging part 2

```python
d1=pd.DataFrame({
    "city" : ["lucknow","kanpur","delhi","pune"],
    "temperature" : [32,45,30,40]
})
```

```python
d1
```

```python
d2=pd.DataFrame({
    "city" : ["pune","lucknow","kanpur","delhi"],
    "humidity" : [62,65,67,70]
})
```

```python
d2
```

```python
df=pd.merge(d1,d2, on='city')
```

```python
df
```

```python
pd.merge(d2,d1,on=['city'], how='outer')
```

```python
pd.merge(d2,d1, on=['city'],how='left')
```
