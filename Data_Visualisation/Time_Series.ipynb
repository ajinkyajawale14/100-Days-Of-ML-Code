# Visualising Time Series Data



```python
df=pd.merge(market_df, orders_df, how='inner',on='Ord_id')
df.head()
```

```python
df.info()
```

```python
#convert object to date type
df['Order_Date']=pd.to_datetime(df['Order_Date'])
df.info()
```

now we can see that
Order_Date             8399 non-null datetime64[ns]

its date time series type

group by Order_Date and compute sum of sales on each day


```python
# aggrgating total sales on each day

time_df=df.groupby('Order_Date')['Sales'].sum()
print(time_df.head())

print(type(time_df))
```

```python
#time series plot

#figure size
plt.figure(figsize=(16,8))

#tsplot
sns.tsplot(data=time_df)
plt.show()
```

# Derived Date Metrics for Visualisation

Derived Variables from Date such as month and year and using them to identify hidden patterns


```python
#extracting moth and year from date

#extract month
df['month'] = df['Order_Date'].dt.month

#extract year 
df['year'] = df['Order_Date'].dt.year

df.head()
```

```python
df_time = df.groupby(["year","month"]).Sales.mean()
df_time.head()
```

```python
plt.figure(figsize=(8,6))
#timeseries plot
sns.tsplot(df_time)
plt.xlabel("Time")
plt.ylabel("Sales")
plt.show()
```

another way to visualising numeric variables such as Sales, across the year and month. we can pivot the month column to create a wide-format dataframe, and then plot a heatmap

```python
# pivoting the data using 'month'
year_month=pd.pivot_table(df,values="Sales",index='year',columns='month', aggfunc='mean')
year_month.head()
```

```python
#figure size
plt.figure(figsize=(12,8))

#heatmap with a color map of choice
sns.heatmap(year_month,cmap="YlGnBu")
plt.show()
```
