```python
#matplotlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# plotting 1d array 
x=np.linspace(5,100,100)
y=np.linspace(10,1000,100)

plt.plot(x,y)
```

```python
#show() to explicitly to display the plot
plt.show()
```

```python
plt.plot([1,4,6,8],[3,8,3,5])    #...[x1,x2,x3,x4][y1,y2,y3,y4]
plt.show()
```

```python
#Axis Labeling

plt.plot(x,y)

#x & y lables, and titles

plt.xlabel("current")
plt.ylabel("Voltage")
plt.title("ohms law")

# define range 
#args:   plt.axis(xmin, xmax, ymin, ymax)

plt.xlim([20,80])
plt.ylim([200,800])
plt.show()
```

```python
# colors and line types
x=np.linspace(0,10,20)
y=x*2

#blue color line type'+'
plt.plot(x,y,'+b')

plt.xlabel("a")
plt.ylabel("b")
plt.title("law")

plt.show()

```

```python
#draw multiple types

x=np.linspace(0,5,10)
y=np.linspace(3,6,10)

#plot 3 curves y,y**2,y**3

plt.plot(x,y,'r-', x,y**2,'b+' ,x,y**3, 'g*')
plt.show()
```

```python
#figures & subplots
x=np.linspace(1,10,100)
y=np.log(x)

#initiate new figure
plt.figure()

#create subplot
#create first subplot with 1 row 2 col
plt.subplot(121)
plt.title("y=log(x)")
plt.plot(x,y)


#create the 2 subplot in figure 1

plt.subplot(122)
plt.title("y=log(x)**2")
plt.plot(x,y**2)
plt.show(x,y)
```

```python
plt.figure(1)
#subplot 1
plt.subplot(2,2,1)
plt.title("linear")
plt.plot(x,x)

#subplot 2
plt.subplot(2,2,2)
plt.title("cubic")
plt.plot(x,x**3)

#subplot 3
plt.figure(2)
plt.subplot(2,2,1)
plt.title("log")
plt.plot(x,np.log(x))

#subplot 4
plt.subplot(2,2,2)
plt.title("exponential")
plt.plot(x,x**2)

plt.show()
```
