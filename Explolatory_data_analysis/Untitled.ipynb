# missing values
        
        
   1. blank string is not missing data
   2. any string say'999' in data is missing value but its not always
   
   thumb rule: we should treat any type of missing data as a universally defined missing data
   
   ###   Rules ###
   
   1. assign any missing data as "NA"
   or change the missing value with mean value.
   but\
   it is advisible to delete the missing data row or col 
   if contains partial missing data then go for mean or a constant value through out the data
   
   2. standardizing data
   
    make sure the data has same unit dimensions(e.g. 1kg rice and 2 lt. milk must not be there
   
    5.775984758994 => 5.77  keep it simple
   
    look at some outlier values(both left right extreme values)
   
    for char data INDIA & ind so make it lower or upper
   
    date in string format so make it to date format
   
   3. unicode issue
   
    utf-8 and other any format of the data must be kept constant through out data
    
    2013-aug OR 01/08/2013 creates error so must be in date format
    DD/MM/YYYY
   
    e.g. temperature in summers t=-20 c is invalid so we have to look into that condition
    
    e.g in bussiness rules delivery date > purchase date 
     if not there it creates problems
