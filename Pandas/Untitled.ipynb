# Pandas Dataframe

```python
#creating dataframe
import numpy as np
```

```python
temp=np.random.randint(low=20, high=100 , size=[20,])
name=np.random.choice(['a','b','c','d'],20)
random = np.random.choice([10,11,12,13,14],20)
```

```python
a=list(zip(temp,name,random))
```

```python
df=pd.DataFrame(data=a,columns=['temp','name','random'])
df
```

```python
type(df)
```
